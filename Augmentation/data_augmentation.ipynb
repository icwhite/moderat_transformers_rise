```python
import pandas as pd
import numpy as np
import re
# load dataset
dataset = "RP-Mod"
source = f"./Datasets/{dataset}-folds.csv"
df = pd.read_csv(source)
train_df = df.loc[df["ten_folds"] < 8]
valid_df = df.loc[df["ten_folds"] >= 8 ]
neg_df = train_df.loc[train_df["label"] == 0]
pos_df = train_df.loc[train_df["label"] == 1]

def make_shorter_examples(df, num_new):
    permutation = np.random.choice(len(df), size=num_new)
    perm_df = df.iloc[permutation]
    new_df = perm_df.apply(lambda row: transform_row(row, get_first_sentence), axis=1)
    return df.append(new_df)

def get_first_sentence(text):
    pattern = r',|\.|!|\?|;'
    lst = re.split(pattern, text)
    return lst[0]

def transform_row(row, text_transform):
    new_row = row.copy()
    new_text = text_transform(row["text"])
    new_row["text"] = new_text
    return new_row



# get relevant folds 0 - 7
# get negative and positive examples
# apply data augmentation 
# remove duplicates
# save 
augmented_df = make_shorter_examples(neg_df, 500)
combined_df = pd.concat([augmented_df, pos_df, valid_df])

# neg_df.apply(get_first_sentence(df["text"]))
```
